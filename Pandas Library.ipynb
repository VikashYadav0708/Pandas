# Series Data Structure

```python
l = [3,4,5,6,7,8,1,2]   #list
a = pd.Series(l)        #pd.Series is used to convert the list in series with their index numbers
print(a)
print()

print("Type of a:",type(a))
print()

print("Dimensions:",a.ndim)
print()

print(a[4])     #getting the data through their index number
```

```python
l = [3,4,5,6,7,8]   #list and can used dictionary or anything else too
a = pd.Series(l, index=['a','s','c','z','e','f'],dtype="float",name="DataStructure in Python") 
#pd.Series is used to convert the list in series with their index numbers and index can be assigned as string,int,etc.
#dtype is used to change the data type of the data we have
#name is used to assign a name to our data
print(a)
print()

print("Type of a:",type(a))
print()

print("Dimensions:",a.ndim)


```

```python
s = pd.Series(12)
print(s)
print(type(s))
```

```python
a = pd.Series(12,index=[1,2,3,4,5,6,7])   #printing 7 elements from the data having 12 elements
print(a)   
```

```python
#We use series in Pandas whereas we can use it in NumPy because broadcasting does not affect Series in Pandas but it does 
#to NumPy arrays

a = pd.Series(12,index=[1,2,3,4,5,6,7]) 
b = pd.Series(12,index=[1,2,3,4]) 
print("Addition of two series:",a + b)
#It will add the series without giving a broadcast error as in NumPy but will return NaN(Not a Num)values for the next data.
#We are using Pandas as it can help us to deal with the missing numbers too.
```

# Merging the DataFrames

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,4],"C":[11,12,13,14]})
pd.merge(df1,df2,on="A")   #pd.merge(dataframe1, dataframe2, common part of merging)
#pd.merge() function is used to merge two DataFrames
#for merging something (id) should be common in both the dataframes.In the above dataframes the A part is common
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[11,12,13,14]})
pd.merge(df1,df2,on="A") 
#if any element of the common data changes then while merging only common part will be printed and merged
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[11,12,13,14]})
pd.merge(df1,df2,how="inner")   #shows no change
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[11,12,13,14]})
pd.merge(df1,df2,how="left")   #left shows the df1 dataframe and checks the missing values while merging
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[11,12,13,14]})
pd.merge(df1,df2,how="right")   #right shows the df2 dataframe and checks the missing values while merging
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[11,12,13,14]})
pd.merge(df1,df2,how="outer")   #outer shows both the df1 & df2 dataframe and checks the missing values while merging
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"C":[11,12,13,14]})
pd.merge(df1,df2,how="outer",indicator=True)   
#outer shows both the df1 & df2 dataframe and checks the missing values while merging
#indicator=True shows that the data is present and merged or not
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"B":[11,12,13,14]})
pd.merge(df1,df2,left_index=True,right_index=True) 
#if both dataframes have same data then it doesn't merge them properly so right_index and left_index=True are used to 
#merge the data 
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[1,2,3,5],"B":[11,12,13,14]})
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=("Roll_no","Name_id"))
#suffixes are used to change the name of the column here
```

# Concat in Pandas
Used to join two dataframes or two series

```python
df1 = pd.Series([1,2,3,4])
df2 = pd.Series([11,12,13,14])
pd.concat([df1,df2])   #while using concat() function the dataframes are passed in a form of a list
```

```python
df_1 = pd.DataFrame({"A":[1,2,3,4],"B":[50,60,70,80]})
df_2 = pd.DataFrame({"A":[5,6,7,8],"B":[11,12,13,14]})
pd.concat([df_1,df_2],axis=0)   #merging with respect to rows if axis=0
```

```python
df_1 = pd.DataFrame({"A":[1,2,3,4],"B":[50,60,70,80]})
df_2 = pd.DataFrame({"A":[5,6,7,8],"C":[11,12,13,14]})
pd.concat([df_1,df_2],axis=1)  #merging with respect to columns if axis=1 
```

```python
df_1 = pd.DataFrame({"A":[1,2,3,4],"B":[50,60,70,80]})
df_2 = pd.DataFrame({"A":[5,6],"C":[11,12]})
pd.concat([df_1,df_2],axis=1)  #merging with respect to columns if axis=1
#if dataFrames have less elements then it will show a missing value or NaN value
```

```python
df_1 = pd.DataFrame({"A":[1,2,3,4],"B":[50,60,70,80]})
df_2 = pd.DataFrame({"A":[5,6],"C":[11,12]})
pd.concat([df_1,df_2],axis=1,join="inner")    #NaN values won't be shown
#join="inner" shows the intersection of the merged data
```

```python
df_1 = pd.DataFrame({"A":[1,2,3,4],"B":[50,60,70,80]})
df_2 = pd.DataFrame({"A":[5,6],"C":[11,12]})
pd.concat([df_1,df_2],axis=1,join="outer")    #NaN values will also be shown
#join="inner" shows the union of the merged data
```

```python
df_1 = pd.DataFrame({"A":[1,2,3,4],"B":[50,60,70,80]})
df_2 = pd.DataFrame({"A":[5,6,7,8],"C":[11,12,13,14]})
pd.concat([df_1,df_2],axis=1,keys=["df_1","df_2"])   
#We can create key of merging data in accordance of merged data
```

# GroupBy Function in Pandas
Used to convert Random Data to Sequential Data

```python
import pandas as pd
a = pd.DataFrame({"Name":["a","b","c","d","a","b","a","c","b"],
                  "S_1":[11,12,14,13,15,16,17,18,20],
                  "S_2":[23,24,25,27,22,21,28,21,29]})
a
```

```python
b = a.groupby("Name")  #random data is converted into a sequence wise group

b
```

```python
for x,y in b:
    print(x)  #groupby's the data and print it's indices sequentially
    print(y)  #the same kind of data is grouped at a single place
    print()
```

```python
b.get_group("a")  #gets the grouped data for a single data at the same place
```

```python
b.min()   #prints the minimum values of a,b,c,d from the data
```

```python
b.max()   #prints the maximum values of a,b,c,d from the data
```

```python
b.mean()   #prints the mean values of a,b,c,d from the data
```

```python
b.std()   #prints the standard deviation 
```

```python
b.var()  #prints the variance
```

# Pivot function in Pandas
Used for arrangement of data

```python
df = pd.DataFrame({"Days":["Mon","Tue","Wed","Thur","Fri","Sat"],
                   "Name":["Vikas","Aman","Abhay","Vikas","Aman","Abhay"],
                   "English":[11,12,13,14,16,16],
                   "Maths":[21,22,23,23,25,26]})

df.pivot(index="Days",columns="Name")  #pivot() is used for arrangement of data
#index and columns are the necessary arguments here
```

```python
df.pivot(index="Days",columns="Name",values="English")  #gets the data of english
```

```python
#pivot_table is used for arithmetic operations
df = pd.DataFrame({"Days":["a","b","a","b","b","a"],
                   "Name":["Vikas","Aman","Aman","Vikas","Aman","Vikas"],
                   "English":[11,12,13,14,16,16],
                   "Maths":[21,22,23,23,25,26]})
df.pivot_table(index="Days",columns="Name",aggfunc="sum")
```

```python
df.pivot_table(index="Days",columns="Name",aggfunc="mean")
```
